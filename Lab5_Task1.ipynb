# **Newton's method**

```
# Defining Function
import numpy as np
import numpy as np
import matplotlib.pyplot as plot

def f(x):
   return np.sin(4*x)*x

# Defining derivative of function
def g(x):
   return np.sin(4*x)+4*np.cos(4*x)*x

# Implementing Newton Method
def newton(x0,e,N):
   print('\n\n*** NEWTON METHOD IMPLEMENTATION ***')
   step = 1
   flag = 1
   x1 = 99
   condition = True
   while condition:
      if g(x0) == 0.0:
         print('Divide by zero error!')
         break
      x1 = x0 - f(x0)/g(x0)
      print('Iteration-%d, x1 = %0.6f and f(x1) = %0.6f' % (step, x1, f(x1)))
      x0 = x1
      step = step + 1
      if step > N:
         flag = 0
         break
      condition = abs(f(x1)) > e
      if flag==1:
         print('\nRequired root is: %0.8f' % x1)
      else:
         print('\nNot Convergent.')
   return x1, step      

startingGuess=[]
root=[]
numberIt=[]

e = 0.000001
# Converting N to integer
N = 1000
x0=np.arange(0.1,5,0.05)
for i in x0 :
  startingGuess.append(i)
  x1 , step = newton(i,e,N)
  root.append(x1)
  numberIt.append(step)
```

```
#Plot part
plot.plot(startingGuess, root, lw=2,  label='Root')
#plot.plot(startingGuess, numberIt, label='Number Iterations')


#Give x axis label 
plot.xlabel('Starting Guess')
#Give y axis label 
plot.ylabel('Amplitude')
plot.grid(True, which='both')
plot.axhline(y=0, color='k')
plot.legend()
plot.show()

#Display 
plot.show
```
