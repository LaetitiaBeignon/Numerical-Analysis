```
import numpy as np
import matplotlib.pyplot as plot

def factorial(n):
  if n<2:
        return 1
  else:
        return n*factorial(n-1)

# Get x values of the sin wave
time = np.arange(-3, 3, 0.1);

#Amplitude of the sin wave is sine of a variable like time
amplitude = np.sin(time)

#Amplitude of the sin wave with Tylor Series
amplitude2 = []
for x in time:
  help = 0
  for n in range(2):
    help = help + ((-1)**n)*((x**(2*n+1))/factorial(2*n+1))
  amplitude2.append(help)


#Plot a sine wave using time and amplitude obtained for the sine wave
plot.plot(time, amplitude)
plot.plot(time, amplitude2)

#Give a title for the sine wave plot
plot.title('Sine wave')

#Give x axis label for the sine wave plot
plot.xlabel('Time')
#Give y axis label for the sine wave plot
plot.ylabel('Amplitude=sin(time)')
plot.grid(True, which='both')
plot.axhline(y=0, color='k')
plot.show()

#Display the sine wave
plot.show
```

```
amplitude3 = amplitude2 - amplitude
plot.plot(time, amplitude3)
#Give a title for the sine wave plot
plot.title('Difference between numpy sin and Tylor approximation')

#Give x axis label for the sine wave plot
plot.xlabel('Time')
#Give y axis label for the sine wave plot
plot.ylabel('Amplitude')
plot.grid(True, which='both')
plot.axhline(y=0, color='k')
plot.show()
plot.show
```
