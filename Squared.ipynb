```
import numpy as np
import matplotlib.pyplot as plot

def factorial(n):
  if n<2:
        return 1
  else:
        return n*factorial(n-1)

# Get x values of the sin wave
time = np.arange(-3, 3, 0.1);

amplitude = np.sin(time)*np.sin(time)+np.cos(time)*np.cos(time)-1

amplitude2 = []
for x in time:
  help = 0
  help2 = 0
  for n in range(2):
    help = help + ((-1)**n)*((x**(2*n+1))/factorial(2*n+1))
    help2 = help2 + ((-1)**n)*((x**(2*n))/factorial(2*n))
  amplitude2.append(help*help+help2*help2-1)



plot.plot(time, amplitude)
plot.plot(time, amplitude2)


plot.title('sin**2+cos**2-1')


plot.xlabel('Time')

plot.ylabel('Amplitude')
plot.grid(True, which='both')
plot.axhline(y=0, color='k')
plot.show()


plot.show
```

```
amplitude3 = amplitude2 - amplitude
plot.plot(time, amplitude3)

plot.title('Difference between numpy and Tylor approximation')


plot.xlabel('Time')

plot.ylabel('Amplitude')
plot.grid(True, which='both')
plot.axhline(y=0, color='k')
plot.show()
plot.show
```
