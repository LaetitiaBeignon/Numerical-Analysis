```
#Lagrange Base

import numpy as np
import matplotlib.pyplot as plt

def pn(x):
  return 2*((x**3-12*x**2+41*x-30)/6)+8*((x**3-10*x**2+29*x-20)/10)+4*((x**3-11*x**2+34*x-24)/-4)+10*((x**3-15*x**2+74*x-120)/-60)


xm =[4, 6, 5, 1]
ym= [2,8,4,10]
xplot=np.linspace(1,7,100)
yplot=pn(xplot)

plt.plot(xplot, yplot, lw=2, label='interpolation')

plt.plot(xm, ym, 'o', label='points')

plt.legend()
plt.show()
```

```
#Newton Base

import numpy as np
import matplotlib.pyplot as plt

def pn(x):
  return 2+3*(x-4)+(x-4)*(x-6)-(1/30)*(x-4)*(x-6)*(x-5)


xm =[4, 6, 5, 1]
ym= [2,8,4,10]
xplot=np.linspace(0,6,100)
yplot=pn(xplot)

plt.plot(xplot, yplot, lw=2, label='interpolation')

plt.plot(xm, ym, 'o', label='points')

plt.legend()
plt.show()
```
