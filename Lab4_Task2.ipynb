```
#Newton Method

import math
import numpy as np
import matplotlib.pyplot as plot

# Defining Function
def f(x):
  return x*math.sin(4*x)

# Defining derivative of function
def g(x):
  return math.sin(4*x) + 4*x*math.cos(4*x)

# Implementing Newton Method
def newton(x0,e, ex, N):
  print('\n\n*** NEWTON METHOD IMPLEMENTATION ***')
  step = 1
  flag = 1
  condition = True
  while condition:
    if g(x0) == 0.0:
      print('Divide by zero error!')
      break
    x1 = x0 - f(x0)/g(x0)
    xhelp=x1-x0
    print('Iteration-%d, x1 = %0.6f and f(x1) = %0.6f' % (step, x1, f(x1)))
    x0 = x1
    step = step + 1
    if step > N:
      flag = 0
      break

    condition1 = abs(f(x1))>e
    condition2 = step<N
    condition3 = abs(xhelp)>ex
    condition = condition1 and condition2 and condition3
    
  if flag==1:
    print('\nRequired root is: %0.8f' % x1)
  else:
    print('\nNot Convergent.')
  return x1

# Input Section
x0 = input('Enter Guess: ')
e = input('Tolerable Error: ')
N = input('Maximum Step: ')
ex= input('Second tolerable error:')

# Converting x0 and e to float
x0 = float(x0)
e = float(e)
ex=float(ex)

# Converting N to integer
N = int(N)

# Starting Newton Method
x1 = newton(x0, e, ex, N)

#Plot part
time = np.arange(5,15,0.1)
amplitude = []
amplitude2 = []

for w in time:
  amplitude.append(f(w))
  amplitude2.append(g(w))

plot.plot(time, amplitude, label='f(x)')
plot.plot(time, amplitude2, label='g(x)')
plot.plot(time, g(x0)*time-g(x0)*x0+f(x0), '-.', color='grey', label='differential')
plot.plot(x0, f(x0), 'o', label='starting guess')
plot.plot(x1, f(x1), 'o', label='root')


#Give x axis label 
plot.xlabel('Time')
#Give y axis label 
plot.ylabel('Amplitude')
plot.grid(True, which='both')
plot.axhline(y=0, color='k')
plot.legend()
plot.show()

#Display 
plot.show
```

```
#Newton Method

import math
import numpy as np
import matplotlib.pyplot as plot

# Defining Function
def f(x):
  return x*math.sin(4*x)

# Defining derivative of function
def g(x):
  return math.sin(4*x) + 4*x*math.cos(4*x)

# Implementing Newton Method
def newton(x0,e, ex, N):
  print('\n\n*** NEWTON METHOD IMPLEMENTATION ***')
  step = 1
  flag = 1
  condition = True
  while condition:
    if g(x0) == 0.0:
      print('Divide by zero error!')
      break
    x1 = x0 - f(x0)/g(x0)
    xhelp=x1-x0
    print('Iteration-%d, x1 = %0.6f and f(x1) = %0.6f' % (step, x1, f(x1)))
    x0 = x1
    step = step + 1
    if step > N:
      flag = 0
      break

    condition1 = abs(f(x1))>e
    condition2 = step<N
    condition3 = abs(xhelp)>ex
    condition = condition1 and condition2 and condition3
    
  if flag==1:
    print('\nRequired root is: %0.8f' % x1)
  else:
    print('\nNot Convergent.')
  return x1

startingGuess=[]
root=[]
numberIt=[]

for i in range(10):

  # Input Section
  x0 = input('Enter Guess: ')
  e = input('Tolerable Error: ')
  N = input('Maximum Step: ')
  ex= input('Second tolerable error:')

  # Converting x0 and e to float
  x0 = float(x0)
  e = float(e)
  ex=float(ex)

  # Converting N to integer
  N = int(N)

  startingGuess.append(x0)
  numberIt.append(N)

  # Starting Newton Method
  x1 = newton(x0, e, ex, N)

  root.append(x1)

#Plot part
plot.plot(startingGuess, root, label='Root')
plot.plot(startingGuess, numberIt, label='Number Iterations')


#Give x axis label 
plot.xlabel('Starting Guess')
#Give y axis label 
plot.ylabel('Amplitude')
plot.grid(True, which='both')
plot.axhline(y=0, color='k')
plot.legend()
plot.show()

#Display 
plot.show
```
